```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data eksperimen baru (tegangan vs suhu)
tegangan = np.array([0, 1, 2, 3, 4, 5, 6]).reshape(-1, 1)  # Input (tegangan dalam volt)
suhu = np.array([29, 27, 31, 38, 48, 59, 67])             # Output (suhu dalam Celsius)

# Training Model
model = LinearRegression()
model.fit(tegangan, suhu)

# Prediksi & Evaluasi
suhu_pred = model.predict(tegangan)
print("Koefisien a (kemiringan garis):", model.coef_[0])
print("Intersep b (titik potong sumbu Y):", model.intercept_)

# Visualisasi
plt.plot(tegangan, suhu, 'bo-', label='Data Asli')         # Titik data asli
plt.plot(tegangan, suhu_pred, 'r--', label='Prediksi ML')  # Garis prediksi
plt.xlabel("Tegangan (V)")
plt.ylabel("Suhu (Celsius)")
plt.title("Prediksi Suhu menggunakan Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```
